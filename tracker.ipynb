```python
import yolo
import cv2
import numpy as np
from PIL import Image

model = yolo.YOLO()

img = "./input_data/test_image.png"
try:
    image = Image.open(img)
except:
    print('Open Error! Try again!')
else:
    r_image = model.detect_image(image)
    cv2.imwrite("./output_data/out.jpg", np.asarray(r_image)[..., ::-1])
    #r_image.show()
```
